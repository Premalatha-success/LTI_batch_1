```python
### Data Preparation
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
###  Replace the missing values with mean/median for numerical columns and mode with categorical columns
```

```python
med_loanamount=df["LoanAmount"].median()
med_loanamount
```

```python
df["LoanAmount"]=df["LoanAmount"].replace(np.nan,med_loanamount)
```

```python
df.isnull().sum()
```

```python
### Do for loan amount term and credit history
```

```python
med_term=df["Loan_Amount_Term"].median()
med_term
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].replace(np.nan,med_term)
```

```python
med_credit=df["Credit_History"].median()
med_credit
```

```python
df["Credit_History"]=df["Credit_History"].replace(np.nan,med_credit)
```

```python
df.isnull().sum()
```

```python
mode_gender=df["Gender"].mode()[0]
mode_gender
```

```python
df["Gender"]=df["Gender"].replace(np.nan,mode_gender)
```

```python
### Do for married,dependents,selfemployed
### Print the missing values->0
```

```python
### Dropping insignificant column/s
df=df.drop("Loan_ID",axis=1)
```

```python
df.shape
```

```python
df.head()
```

```python
### Encoding
```

```python
from sklearn.preprocessing import LabelEncoder

# Select categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
Y=df[["Loan_Status"]]
X=df.drop("Loan_Status",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)
```

```python
model_lr.score(X_train,Y_train)
```

```python
model_lr.score(X_test,Y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier()
model_knn.fit(X_train,Y_train)
```

```python
model_knn.score(X_train,Y_train)
```

```python
model_knn.score(X_test,Y_test)
```
